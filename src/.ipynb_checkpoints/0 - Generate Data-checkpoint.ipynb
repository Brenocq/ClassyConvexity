```python
import matplotlib.pyplot as plt
import random
import numpy as np
from keras.preprocessing.image import ImageDataGenerator, array_to_img, img_to_array, load_img

# Create transformations of each class
for classNumber in range(0,6):
    plt.figure(classNumber)
    
    #---------- Define transformations ----------#
    datagen = ImageDataGenerator(rotation_range=90, 
                                 width_shift_range=0.1, 
                                 height_shift_range=0.1, 
                                 zoom_range=0, 
                                 vertical_flip=True,
                                 horizontal_flip=True, 
                                 fill_mode="nearest")

    #---------- Load image ----------#
    img_path = "../Data/aircraft-"+str(classNumber)+"/0.jpg"
    img = load_img(img_path)
    plt.imshow(img)
    plt.axis("off")
    plt.title("Original ", fontsize=16)

    img = img_to_array(img)
    img = img.reshape((1,) + img.shape)

    # Number of images that will generate for each class
    n_augmentations = 100
    plt.figure(figsize=(15, 6))
    i = 1
    
    #---------- Generate new images ----------#
    for batch in datagen.flow(img, 
                              batch_size=1, 
                              seed=21):

        plt.subplot(10, int(np.ceil(n_augmentations * 1./10)), i + 1)
        plt.imshow(array_to_img(batch[0]))
        plt.axis("off")
        plt.suptitle("Augmented ", fontsize=16)    

        augImg = array_to_img(batch[0]);
        path = "../Data/aircraft-"+str(classNumber)+"/"+str(i)+".jpg"
        augImg.save(path)

        i += 1
        if i >= n_augmentations:
            break
```
